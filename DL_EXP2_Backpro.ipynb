```python
import numpy as np
```

```python
def activation_func(net):
    return 1 / (1 + np.exp(-net))
```

```python
def calc_error(output, det_output):
    return (1/2)*(det_output - output)**2
```

#### Backpropogation

```python
def forward_pass(x_input, weights, biases):
    net_h0 = (x_input[0] * weights[0]) + (x_input[1] * weights[1]) + biases[0]

    outh0 = activation_func(net_h0)


    net_h1 = (x_input[0] * weights[2]) + (x_input[1] * weights[3]) + biases[1]

    outh1 = activation_func(net_h1)


    net_o0 = (outh0 * weights[4]) + (outh1 * weights[5]) + biases[2]

    out_o0 = activation_func(net_o0)

    return outh0, outh1, out_o0
```

```python
def backward_pass(x_input, weights, biases, desired_output, learning_rate, outh0, outh1, out_o0):
    delta_o0 = (desired_output - out_o0) * out_o0 * (1 - out_o0)

    delta_h0 = delta_o0 * weights[4] * outh0 * (1 - outh0)

    delta_h1 = delta_o0 * weights[5] * outh1 * (1 - outh1)

    dw0 = learning_rate * delta_h0 * x_input[0]
    dw1 = learning_rate * delta_h0 * x_input[1]
    dw2 = learning_rate * delta_h1 * x_input[0]
    dw3 = learning_rate * delta_h1 * x_input[1]
    dw4 = learning_rate * delta_o0 * outh0
    dw5 = learning_rate * delta_o0 * outh1
    dw_values = [dw0, dw1, dw2, dw3, dw4, dw5]

    db0 = learning_rate * delta_h0
    db1 = learning_rate * delta_h1
    db2 = learning_rate * delta_o0
    db_values = [db0, db1, db2]

    return dw_values, db_values
```

```python
X = [1,0]
w = [0.5, -0.3, 0.8, 0.2, 0.6, -0.5]
b = [0.2, -0.4, 0.3]
d = [1]
learning_rate = 0.01
```

```python
epochs = 10000
error_threshold = 0.001

X = [[1, 0]]
d = [1]

w = [0.5, -0.3, 0.8, 0.2, 0.6, -0.5]
b = [0.2, -0.4, 0.3]
learning_rate_initial = 0.5

def train(weights, biases, learning_rate, epochs, error_threshold):
    current_w = list(weights)
    current_b = list(biases)

    for epoch in range(epochs):
        total_epoch_error = 0

        for i in range(len(X)):
            x_input = X[i]
            desired_output = d[i]

            outh0, outh1, out_o0 = forward_pass(x_input, current_w, current_b)

            error = calc_error(out_o0, desired_output)
            total_epoch_error += error

            dw_values, db_values = backward_pass(x_input, current_w, current_b, desired_output, learning_rate, outh0, outh1, out_o0)

            for j in range(len(current_w)):
                current_w[j] += dw_values[j]

            for j in range(len(current_b)):
                current_b[j] += db_values[j]

        if epoch % 1000 == 0 or total_epoch_error < error_threshold:
            print(f"Epoch {epoch}: Total Error = {total_epoch_error:.6f}")
            print(f"  Output for input {X[0]}: {forward_pass(X[0], current_w, current_b)[2]:.4f}")

        if total_epoch_error < error_threshold:
            print(f"Stopping condition met at epoch {epoch+1}. Total Error ({total_epoch_error:.6f}) <= Threshold ({error_threshold:.6f})")
            break

    return current_w, current_b

updated_w, updated_b = train(w, b, learning_rate_initial, epochs, error_threshold)

print(f"\nFinal updated weights:")
for w in updated_w:
    print(f"\t{w:.4f}")
print(f"Final updated biases:")
for b in updated_b:
    print(f"\t{b:.4f}")
```

#### On XOR input with SGD

```python
# XOR input
x1 = [0, 0]
x2 = [0, 1]
x3 = [1, 0]
x4 = [1, 1]
y = [0, 1, 1, 0]

epochs = 2000
error_threshold = 0.001

X = [x1, x2, x3, x4]
d = y

w = [0.5, -0.3, 0.8, 0.2, 0.6, -0.5]
b = [0.2, -0.4, 0.3]
learning_rate_initial = 0.01
```

```python
def train_SGD(X_data, y_data, epochs_count, lr_val, initial_weights_list, initial_biases_list):
    w1, w2, w3, w4, w5, w6 = (
        initial_weights_list[0],
        initial_weights_list[1],
        initial_weights_list[2],
        initial_weights_list[3],
        initial_weights_list[4],
        initial_weights_list[5],
    )
    b1, b2, b3 = initial_biases_list[0], initial_biases_list[1], initial_biases_list[2]

    for epoch in range(epochs_count):
        total_loss = 0
        for i in range(len(X_data)):
            x_input = X_data[i]
            target = y_data[i]

            current_weights = [w1, w2, w3, w4, w5, w6]
            current_biases = [b1, b2, b3]

            outh0, outh1, out_o0 = forward_pass(x_input, current_weights, current_biases)

            current_loss = calc_error(out_o0, target)
            total_loss += current_loss

            dw_values, db_values = backward_pass(
                x_input,
                current_weights,
                current_biases,
                target,
                lr_val,
                outh0,
                outh1,
                out_o0,
            )

            w1 += dw_values[0]
            w2 += dw_values[1]
            w3 += dw_values[2]
            w4 += dw_values[3]
            w5 += dw_values[4]
            w6 += dw_values[5]

            b1 += db_values[0]
            b2 += db_values[1]
            b3 += db_values[2]

        avg_loss = total_loss / len(X_data)

        print(f"Epoch {epoch+1}: Avg. Loss = {avg_loss}")

        # if epoch % 1000 == 0:
        #     print(f"Epoch {epoch}: Total Error = {total_loss:.6f}")
        #     for input_idx in range(len(X_data)):
        #         current_input = X_data[input_idx]
        #         output_for_input = forward_pass(
        #             current_input, [w1, w2, w3, w4, w5, w6], [b1, b2, b3]
        #         )[2]
        #         print(
        #             f"  Output for input {current_input}: {output_for_input:.4f} (Desired: {y_data[input_idx]}) "
        #         )

        if avg_loss < error_threshold:
            print(
                f"Stopping condition met at epoch {epoch+1}. Total Error ({total_loss:.6f}) <= Threshold ({error_threshold:.6f})"
            )
            break

    print("\nFinal updated weights:")
    print(f"\t{w1:.4f}")
    print(f"\t{w2:.4f}")
    print(f"\t{w3:.4f}")
    print(f"\t{w4:.4f}")
    print(f"\t{w5:.4f}")
    print(f"\t{w6:.4f}")
    print("Final updated biases:")
    print(f"\t{b1:.4f}")
    print(f"\t{b2:.4f}")
    print(f"\t{b3:.4f}")

    return [w1, w2, w3, w4, w5, w6], [b1, b2, b3]


print("Training for XOR dataset:")
updated_w_xor, updated_b_xor = train_SGD(
    X,
    d,
    epochs,
    learning_rate_initial,
    w,
    b,
)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
```

```python
import kagglehub

path = kagglehub.dataset_download("yasserh/titanic-dataset")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv(f"{path}/Titanic-Dataset.csv")
df
```

```python
df.describe()
```

```python
y_titanic = df['Survived']
X_titanic = df.drop(columns=["Survived", "Name", "Ticket", "Cabin"])
```

```python
X_titanic["Sex"] = X_titanic["Sex"].map({"male": 0, "female": 1})
```

```python
X_titanic = pd.get_dummies(X_titanic, columns=["Embarked"], drop_first=True)
```

```python
X_titanic = X_titanic.fillna(X_titanic.mean())
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_titanic)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_titanic, test_size=0.2, random_state=42
)
```

```python
mlp = MLPClassifier(
    hidden_layer_sizes=(10,),
    activation="logistic",
    solver="sgd",
    learning_rate_init=0.01,
    random_state=42,
    max_iter=2000
)
```

```python
mlp.fit(X_train, y_train)
```

```python
print("\nWeights (Coefficients):")
for i, coef in enumerate(mlp.coefs_):
    print(f"\nLayer {i} weights:\n", coef)

print("\nBiases (Intercepts):")
for i, intercept in enumerate(mlp.intercepts_):
    print(f"\nLayer {i} biases:\n", intercept)
```
