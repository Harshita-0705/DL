```python
import numpy as np
```

```python
x = np.array([
    [1, 0, 1, 1],
    [0, 0, 0, 1],
    [1, 0, 0, 1],
    [1, 1, 1, 0]
])
```

```python
w = np.array([
    [0.9, 0.3, 0.1, 0.2],
    [0.1, 0.6, 0.2, 0.1]
])
```

```python

# Learning rate
lr = 0.1

# Number of epochs
epochs = 2

for e in range(epochs):

    print(f"\nEpoch {e+1}")

    for i in range(len(x)):

        print("\nInput Vector:", x[i])

        # Calculate Euclidean Distance
        distances = np.sum((w - x[i])**2, axis=1)

        print("Distances:", distances)

        # Find Best Matching Unit (BMU)
        bmu = np.argmin(distances)

        print("Winning Neuron:", bmu + 1)

        # Update winning neuron's weights
        w[bmu] = w[bmu] + lr * (x[i] - w[bmu])

        print("Updated Weights:")
        print(w)

    # Reduce learning rate
    lr = lr * 0.5

print("\nFinal Weights:")
print(w)
```
